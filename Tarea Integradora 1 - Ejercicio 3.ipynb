```python
#Son eventos dependientes, la cantidad de eventos en una determinada provincia esta en funcion del año
#en que se analice
#Traemos las librerias necesarias y el set de datos

%matplotlib inline
import pandas
import numpy
import seaborn

seaborn.set_style('whitegrid')
seaborn.set_context('talk')

#Cargamos en una variable llamada "Dataset" los datos del csv de violencia
dataset = pandas.read_csv('https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/violencia-institucional-2018-01.csv', encoding='utf8')

import six
def convert_to_int(row):
    try:
        return float(row)
    except ValueError:
        if isinstance(row, six.string_types):
            return float(row.split(',')[0])
        return row

dataset.loc[:,'year'] = dataset[u'año'].apply(convert_to_int)
```

```python
#Tome las variables. y=año y x=provincia
#con este codigo veo la tabla cruzada como cantidad de casos por cada cruce

tablaCruzada = pandas.crosstab(index=dataset["año"], columns=dataset["provincia"])

tablaCruzada
```

```python
#Tome las variables. año y provincia
#con este codigo veo la tabla cruzada como % de frecuencia

tablaCruzada = pandas.crosstab(index=dataset["año"].dropna(), columns=dataset["provincia"].dropna()).apply(lambda r: r/len(dataset) *100, axis=1)
tablaCruzada
```

```python
#seaborn.heatmap(tablaCruzada, annot=True, vmin=0, vmax=100)

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,10))         # Sample figsize in inches
seaborn.heatmap(tablaCruzada.iloc[:, 1:100:], annot=True, linewidths=.5, ax=ax)
seaborn.set(font_scale=0.9)

#seaborn.heatmap(tablaCruzada, cmap="YlGnBu")
#seaborn.heatmap(tablaCruzada, cmap="Blues")
#seaborn.heatmap(tablaCruzada, cmap="BuPu")
#seaborn.heatmap(tablaCruzada, cmap="Greens")
```
