## 2.3. Codificar variables

Es necesario que las variables categóricas sean etiquetadas como variables numéricas, no como cadenas




```python
from sklearn import preprocessing
permisos.describe(include=['O'])
```

```python
column='Street_Name'

NuevaStreet = preprocessing.LabelEncoder()
NuevaStreet.fit(permisos[column])
```

```python
dict(zip(NuevaStreet.classes_,NuevaStreet.transform(NuevaStreet.classes_)))
```

Aqui vemos la clasificacion de las calles

```python
Aplicamos la codificacion con pandas
```

```python
permisos[column] = NuevaStreet.transform(permisos[column]) 
permisos.head()
```

```python
#Volvemos atras


permisos[column] = NuevaStreet.inverse_transform(permisos[column]) 
permisos.head()
```

## Identificar y documentar valores atípicos

```python
##Antes que nada vamos a seleccionar una variable. Por ejemplo, Unidades

#outliers = permisos[permisos.Unit > (permisos.Unit.mean() + 3 * permisos.Unit.std())]
#display(outliers.describe())

x = permisos["Unit"].dropna()  #utilizo dropna() para sacar los valores NULOS del calculo dado que da error en la generacion del grafico
plt.hist(x)
plt.ylabel('No of times')
plt.show()
```

```python
#Ahora al analizar el grafico, aplico 
valores = permisos["Unit"].dropna()

mean = np.mean(valores, axis=0)
sd = np.std(valores, axis=0)

final_list = [x for x in valores if (x > mean - 2 * sd)]
final_list = [x for x in final_list if (x < mean + 2 * sd)]
##print(final_list)

plt.hist(final_list)
plt.ylabel('No of times')
plt.show()
```

Con este codigo logro quitar aquellos valores que parecen ser atipicos y que pueden afectar a mi analisis. Como atipico se tomo los valores que se desvian mas de 3 veces la desviacion estandar de la variable
