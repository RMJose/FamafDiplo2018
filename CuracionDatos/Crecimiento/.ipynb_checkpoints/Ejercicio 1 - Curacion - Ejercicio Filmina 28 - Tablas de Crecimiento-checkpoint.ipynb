```python
#Importamos Libreria y llenamos dataset

import matplotlib.pyplot as plt
import pandas as pd
datosPeso = pd.read_csv('https://www.cdc.gov/growthcharts/data/zscore/wtageinf.csv')
datosPeso = datosPeso[['Sex','Agemos','P3','P5','P10','P25','P50','P75','P90','P95']]
datosPeso.describe(include='all')





```

```python
datosPesoSexoÑiños = datosPeso.loc[datosPeso['Sex'] == 1]

plt.figure(figsize=(20,8))
plt.ylabel('Peso')
plt.xlabel('Edad (meses de vida)')
plt.title('Peso segun Edad - Niños')


#datosGraficos = datosPesoSexoÑiños
num=0
for column in datosPesoSexoÑiños.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(datosPesoSexoÑiños["Agemos"], datosPesoSexoÑiños[column], marker='',  linewidth=3, alpha=0.9, label=column)
        

plt.legend(loc=2, ncol=1)
```

```python
#Hacemos lo mismo ahora para las niñas
```

```python
datosPesoSexoNiñas = datosPeso.loc[datosPeso['Sex'] == 2]

plt.figure(figsize=(20,8))
plt.ylabel('Peso')
plt.xlabel('Edad (meses de vida)')
plt.title('Peso segun Edad - Niñas')


#datosGraficos = datosPesoSexoNiñas
num=0
for column in datosPesoSexoNiñas.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(datosPesoSexoNiñas["Agemos"], datosPesoSexoNiñas[column], marker='',  linewidth=3, alpha=0.9, label=column)
        

plt.legend(loc=2, ncol=1)
```

### Realizamos el mismo analisis, pero ahora para la altura

```python
datosAltura = pd.read_csv('lenageinf.csv')
##datosAltura = pd.read_csv('https://www.cdc.gov/growthcharts/data/zscore/lenageinf.csv')
datosAltura = datosAltura[['Sex','Agemos','P3','P5','P10','P25','P50','P75','P90','P95']]


datosAlturaSexoNiños = datosAltura.loc[datosAltura['Sex'] == "1"]
datosAlturaSexoNiños.describe(include='all')
```

```python

plt.figure(figsize=(20,8))
plt.ylabel('Altura')
plt.xlabel('Edad (meses de vida)')
plt.title('Altura segun Edad - Niños')


num=0
for column in datosAlturaSexoNiños.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(pd.to_numeric(datosAlturaSexoNiños["Agemos"], errors='ignore') , pd.to_numeric(datosAlturaSexoNiños[column], errors='ignore'), marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=1)
```

```python
##Ahora para las mujeres

datosAlturaSexoNiñas = datosAltura.loc[datosAltura['Sex'] == "2"]
datosAlturaSexoNiñas.describe(include='all')
```

```python
plt.figure(figsize=(20,8))
plt.ylabel('Altura')
plt.xlabel('Edad (meses de vida)')
plt.title('Altura segun Edad - Niños')


num=0
for column in datosAlturaSexoNiñas.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(pd.to_numeric(datosAlturaSexoNiñas["Agemos"], errors='ignore') , pd.to_numeric(datosAlturaSexoNiñas[column], errors='ignore'), marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=1)
```
