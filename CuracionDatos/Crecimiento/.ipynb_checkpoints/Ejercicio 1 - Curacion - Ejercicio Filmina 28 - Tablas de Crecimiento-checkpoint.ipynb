```python
#Importamos Libreria y llenamos dataset

import matplotlib.pyplot as plt
import pandas as pd
datosPeso = pd.read_csv('https://www.cdc.gov/growthcharts/data/zscore/wtageinf.csv')
datosPeso = datosPeso[['Sex','Agemos','P3','P5','P10','P25','P50','P75','P90','P95']]
datosPeso.describe(include='all')





```

```python
datosPesoSexoios = datosPeso.loc[datosPeso['Sex'] == 1]

plt.figure(figsize=(20,8))
plt.ylabel('Peso')
plt.xlabel('Edad (meses de vida)')
plt.title('Peso segun Edad - Nios')


#datosGraficos = datosPesoSexoios
num=0
for column in datosPesoSexoios.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(datosPesoSexoios["Agemos"], datosPesoSexoios[column], marker='',  linewidth=3, alpha=0.9, label=column)
        

plt.legend(loc=2, ncol=1)
```

```python
#Hacemos lo mismo ahora para las nias
```

```python
datosPesoSexoNias = datosPeso.loc[datosPeso['Sex'] == 2]

plt.figure(figsize=(20,8))
plt.ylabel('Peso')
plt.xlabel('Edad (meses de vida)')
plt.title('Peso segun Edad - Nias')


#datosGraficos = datosPesoSexoNias
num=0
for column in datosPesoSexoNias.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(datosPesoSexoNias["Agemos"], datosPesoSexoNias[column], marker='',  linewidth=3, alpha=0.9, label=column)
        

plt.legend(loc=2, ncol=1)
```

### Realizamos el mismo analisis, pero ahora para la altura

```python
datosAltura = pd.read_csv('lenageinf.csv')
##datosAltura = pd.read_csv('https://www.cdc.gov/growthcharts/data/zscore/lenageinf.csv')
datosAltura = datosAltura[['Sex','Agemos','P3','P5','P10','P25','P50','P75','P90','P95']]


datosAlturaSexoNios = datosAltura.loc[datosAltura['Sex'] == "1"]
datosAlturaSexoNios.describe(include='all')
```

```python

plt.figure(figsize=(20,8))
plt.ylabel('Altura')
plt.xlabel('Edad (meses de vida)')
plt.title('Altura segun Edad - Nios')


num=0
for column in datosAlturaSexoNios.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(pd.to_numeric(datosAlturaSexoNios["Agemos"], errors='ignore') , pd.to_numeric(datosAlturaSexoNios[column], errors='ignore'), marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=1)
```

```python
##Ahora para las mujeres

datosAlturaSexoNias = datosAltura.loc[datosAltura['Sex'] == "2"]
datosAlturaSexoNias.describe(include='all')
```

```python
plt.figure(figsize=(20,8))
plt.ylabel('Altura')
plt.xlabel('Edad (meses de vida)')
plt.title('Altura segun Edad - Nios')


num=0
for column in datosAlturaSexoNias.drop('Sex', axis=1):
    num+=1
    if num>1:
        plt.plot(pd.to_numeric(datosAlturaSexoNias["Agemos"], errors='ignore') , pd.to_numeric(datosAlturaSexoNias[column], errors='ignore'), marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=1)
```
